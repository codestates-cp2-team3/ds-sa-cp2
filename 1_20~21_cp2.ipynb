### 2. 사용자 기반 협업 필터링(코사인 유사도)

  - 추천 대상인 사용자와 유사도가 높은 사용자일수록, 취향을 정확하게 반영할 수 있다.

```python
user_data=df_c.pivot_table('rating',index='user_id',columns='movie_id').fillna(0)
user_data
```

2. 비슷한 성향의 유저 찾기

```python
def similar_user(user_id, matrix, k=10):
  # 현재 유저에 대한 데이터 프레임 만들기
  user = matrix[matrix.index == user_id]

  #일치하지 않는 다른 유저
  other_users = matrix[matrix.index != user_id]

  # 현재 유저와 다른 유저와의 코사인 유사도 계산(유사도, p-value)
  # list 변환
  similarites = cosine_similarity(user,other_users)[0].tolist()

  # 다른 유저의 인덱스 목록 생성
  other_users_list = other_users.index.tolist()

  # 인덱스/유사도로 이루어진 딕셔너리 생성
  user_similarity = dict(zip(other_users_list,similarites))

  # 딕셔너리 정렬
  # key = operator.itemgetter(1) -> value값으로 오름차순
  sort_user_similarity = sorted(user_similarity.items(), key=operator.itemgetter(1))
  sort_user_similarity.reverse() #내림차순

  # 가장 높은 유사도 k개 정렬
  top_users_similarities = sort_user_similarity[:k]
  users = [i[0] for i in top_users_similarities]

  return users
```

```python
def recommend_item(user_index, similar_user_indices, matrix, items=10):
  # 유저와 비슷한 유저 가져오기
  # similar_user_indices = similar_user(1,user_data,10)
  similar_users = matrix[matrix.index.isin(similar_user_indices)]
  
  # 비슷한 유저 평균 (계산 영화에 대한 평균)
  mean_similar_users= similar_users.mean(axis=0)
  # 데이터프레임
  df_similar_users = pd.DataFrame(mean_similar_users, columns=['user_similarity'])

  # 현재 유저의 벡터 
  user_df = matrix[matrix.index==user_index]

  # 현재 유저의 영화평점에 대한 데이터 정렬
  trans_user_df = user_df.T
  # 컬럼명 변경
  trans_user_df.columns=['rating']

  # 미시청 영화 만들기
  trans_user_df = trans_user_df[trans_user_df['rating']==0]
  movie_unseen = trans_user_df.index.tolist()

  # 안본 영화 필터링
  similar_users_filter = df_similar_users[df_similar_users.index.isin(movie_unseen)]

  # 비슷한 유저 평균 기준으로 내림차순
  order_similar_user = similar_users_filter.sort_values(by=['user_similarity'], ascending=False)

  # 상위 10개의 값 가져오기
  top_n_movie = order_similar_user.head(items)
  top_n_movie_indices = top_n_movie.index.tolist()

  # movie_df에서 top10개 찾기
  movie_title =movie_df[movie_df['movie_id'].isin(top_n_movie_indices)]['movie_title']
  
  return movie_title
```

```python
# 비슷한 10명의 유저 찾기
similar_user_indices = similar_user(369, user_data, 10)

#추천 10개의 영화
recommend_movie = recommend_item(369, similar_user_indices, user_data, 10)
```
